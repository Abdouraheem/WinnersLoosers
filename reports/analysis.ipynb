```python
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
#Import commodity prices into dataframe from CSV
commodity_price_df = pd.read_csv(
    Path('../../data/rawdata/commodity_prices.csv'),
    parse_dates=True,
    infer_datetime_format=True
)

commodity_price_df['Date'] = pd.to_datetime(commodity_price_df['Date'])
commodity_price_df['Date'] = commodity_price_df['Date'].apply(lambda x: x.strftime('%Y-%m-%d'))
commodity_price_df.rename(columns = {'Date':'time'}, inplace=True)

commodity_price_df = commodity_price_df.set_index('time')

#Check dataframe and data types

display(commodity_price_df.head())
commodity_price_df.dtypes
```

```python
commodity_daily_returns = commodity_price_df.pct_change().dropna()
```

```python
commodity_daily_returns.head()
```

```python
equity_price_df = pd.read_csv(
    Path('../../data/rawdata/equity_prices.csv'),
    parse_dates=True, 
    infer_datetime_format=True
)


#Check dataframe and data types

display(equity_price_df.head())
equity_price_df.dtypes

equity_price_df['time'] = pd.to_datetime(equity_price_df['time'])
```

```python
equity_price_df['time'] = equity_price_df['time'].apply(lambda x: x.strftime('%Y-%m-%d'))
```

```python
equity_price_df = equity_price_df.set_index('time')
```

```python
equity_price_df.head()
```

```python
#Determine if any columns are missing large amounts of data, such as new IPO
equity_price_df.isnull().sum().sort_values().tail(20)
```

```python
#Remove columns with missing information, likely new IPO's
equity_price_df = equity_price_df.dropna(axis=1)
equity_price_df.isnull().sum().sort_values()
```

```python
equity_daily_returns = equity_price_df.pct_change().dropna()
display(equity_daily_returns.head())
display(equity_daily_returns.tail())
```

```python
oil_correlation_df = pd.DataFrame()

oil_correlation_df = pd.concat([commodity_daily_returns['oil'], equity_daily_returns], axis='columns', join='inner')
oil_correlation_df.isnull().sum()
```

```python
oil_correlation_df = oil_correlation_df.corr(method='pearson')
```

```python
oil_correlation_df = oil_correlation_df.loc['oil']
```

```python
oil_correlation_df = oil_correlation_df.sort_values(ascending=False)
display(oil_correlation_df.head())
display(oil_correlation_df.tail())
```
