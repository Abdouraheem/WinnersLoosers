```python
import os
import requests
import pandas as pd
from dotenv import load_dotenv
from pathlib import Path
import alpaca_trade_api as tradeapi
```

```python
load_dotenv()
```

```python
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

display(type(alpaca_api_key))
display(type(alpaca_secret_key))
```

```python
alpaca = tradeapi.REST(
     alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
nasdaq_df = pd.read_csv(
    Path('rawdata/nasdaq_screener.csv')
)
```

```python
nasdaq_df = nasdaq_df.sort_values(by=['Market Cap'], ascending=False)
nasdaq_df = nasdaq_df.reset_index(drop=True)
```

```python
nasdaq_filtered_100 = nasdaq_df.iloc[:100]
nasdaq_filtered_200 = nasdaq_df.iloc[100:200]
nasdaq_filtered_300 = nasdaq_df.iloc[200:300]
nasdaq_filtered_400 = nasdaq_df.iloc[300:400]
nasdaq_filtered_500 = nasdaq_df.iloc[400:500]
```

```python
tickers_100 = nasdaq_filtered_100['Symbol']
tickers_200 = nasdaq_filtered_200['Symbol']
tickers_300 = nasdaq_filtered_300['Symbol']
tickers_400 = nasdaq_filtered_400['Symbol']
tickers_500 = nasdaq_filtered_500['Symbol']
```

```python
start_date = pd.Timestamp("2020-10-31", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2021-11-30", tz="America/New_York").isoformat()
timeframe = "1D"
limit_rows = 1000
```

```python
prices_df_100 = alpaca.get_barset(
    tickers_100,
    timeframe,
    start=start_date,
    end=end_date,
    limit=limit_rows
).df

prices_df_100
```

```python
prices_df_200 = alpaca.get_barset(
    tickers_200,
    timeframe,
    start=start_date,
    end=end_date,
    limit=limit_rows
).df
```

```python
prices_df_300 = alpaca.get_barset(
    tickers_300,
    timeframe,
    start=start_date,
    end=end_date,
    limit=limit_rows
).df
```

```python
prices_df_400 = alpaca.get_barset(
    tickers_400,
    timeframe,
    start=start_date,
    end=end_date,
    limit=limit_rows
).df
```

```python
prices_df_500 = alpaca.get_barset(
    tickers_500,
    timeframe,
    start=start_date,
    end=end_date,
    limit=limit_rows
).df
```

```python
equity_prices = pd.concat([prices_df_100, prices_df_200, prices_df_300, prices_df_400, prices_df_500])
equity_prices.head()
```
